```python
def elbow_method(X_scaled):
    # Elbow method to find the optimal number of clusters
    wcss = []
    for i in range(1, 11):
        kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
        kmeans.fit(X_scaled)
        wcss.append(kmeans.inertia_)
    return wcss
```

```python
def plot_elbow(wcss,point=0):
    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.plot(range(1, 11), wcss, marker='o', linestyle='-', color='#44AA99', markersize=8)
    plt.title('Elbow Method for Optimal Number of Clusters', fontsize=15, fontweight='bold')
    plt.xlabel('Number of Clusters', fontsize=12)
    plt.ylabel('WCSS', fontsize=12)
    plt.xticks(range(1, 11), fontsize=10)
    plt.yticks(fontsize=10)
    plt.grid(True, linestyle='--', linewidth=0.7)
    if point != 0:
        plt.axvline(x=point, color='r', linestyle='--')  # Optional: Highlight the elbow point if known

    plt.tight_layout()
```

**1-Grouping using Age and Annual Income(k$)**

```python
# Extract features
X1 = data[['Age', 'Annual Income (k$)']]

# Standardize the data
scaler = StandardScaler()
X1_scaled = scaler.fit_transform(X1)

# Find the WCSS For the Elbow Method
wcss1 = elbow_method(X1_scaled)

# Plot the Result Of the Elbow Method
plot_elbow(wcss1,4)
```

```python
# Apply KMeans clustering with 4 clusters
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=42)
y_kmeans1 = kmeans.fit_predict(X1_scaled)

# Create a meshgrid of points to plot the decision boundaries
x_min, x_max = X1_scaled[:, 0].min() - 1, X1_scaled[:, 0].max() + 1
y_min, y_max = X1_scaled[:, 1].min() - 1, X1_scaled[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

# Predict the cluster for each point in the meshgrid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundaries
plt.figure(figsize=(10, 7))
plt.contourf(xx, yy, Z, alpha=0.4, cmap=ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF', '#FFD700']))

# Scatter the actual clusters
scatter_clusters(y_kmeans1, X1_scaled, X1, 4)

# Plot centroids of the clusters
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', label='Centroids', marker='X')

# Add plot title and legend
plt.title("Customer Clusters")
plt.legend()
plt.show()
```

**Now we will try to Apply PCA To the data and Then Use KMean**

```python
#### Prepare the Data ####
# Convert categorical 'Gender' to numerical
data['Gender'] = data['Gender'].map({'Male': 0, 'Female': 1})

# Extract features
X = data[['Age', 'Gender', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Show the Shape Of the Features Before Applying PCA
X_scaled.shape
```

```python
# Apply PCA to reduce dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Show the Features Shape after applying Dimentionality Reduction(PCA)
X_pca.shape
```

```python
# Find the The Suitable number of clusters using Elbow Method

# Find the WCSS For the Elbow Method
wcss = elbow_method(X_pca)

# Plot the Result Of the Elbow Method
plot_elbow(wcss,4)
```

```python
# Apply KMeans clustering with 4 clusters
kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=42)
y_kmeans = kmeans.fit_predict(X_pca)

# Create a meshgrid of points for decision boundaries
x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.001),  # Increased resolution
                     np.arange(y_min, y_max, 0.001))

# Predict the cluster for each point in the meshgrid
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot the decision boundaries with adjusted levels
boundary_colors = ['#FFAAAA', '#AAFFAA', '#AAAAFF', '#FFD700']
plt.figure(figsize=(10, 7))
plt.contourf(xx, yy, Z, alpha=0.3, cmap=ListedColormap(boundary_colors), levels=np.arange(-0.5, 4.5, 1))

# Scatter the actual clusters from X_pca
scatter_clusters(y_kmeans, X_pca, X, 4)

# Plot centroids of the clusters
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='black', label='Centroids', marker='X')

# Add plot title and legend
plt.title("KMeans Clustering with Decision Boundaries (PCA Transformed Data)")
plt.legend()
plt.show()
```
